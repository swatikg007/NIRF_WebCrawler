```python
import requests as req
from bs4 import BeautifulSoup
import re
import os
```

```python
page = requests.get("https://www.nirfindia.org/2021/ManagementRanking.html")
soup = BeautifulSoup(page.content,"html.parser")
print(soup)
```

```python
data_jpgs = []
data_pdfs = []
for link in soup.find_all("a"):
    data = link.get("href")
    # print(isinstance(data, str))
    if data is not None and isinstance(data, str):
        if data.split(".")[-1] == "jpg":
            data_jpgs.append(data)
        elif  data.split(".")[-1] == "pdf":
            data_pdfs.append(data)
    # print(data)
    # print("-" * 80)
```

```python
print(data_jpgs)
print(len(data_jpgs))
print(data_pdfs)
print(len(data_pdfs))
```
